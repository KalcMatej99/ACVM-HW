```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
for n in ["s.csv", "bins.csv", "alfa.csv", "stop.csv"]:
    print(n)
    df = pd.read_csv(n, sep = " ")
    df.plot(x = "v", y = "fps")
    df.plot(x = "v", y = "f", legend=None)
    plt.xlabel("Number of bins")
    plt.ylabel("Number of fails")
    plt.title("Number of fails w.r.t. number of bins")

    plt.show()
```
