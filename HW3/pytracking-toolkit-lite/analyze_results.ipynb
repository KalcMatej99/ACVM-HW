```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("rez.csv")

```

```python
df.head()
```

```python
for col in ["alfa", "sigma", "enlarge_factor"]:
    unique_sigmas = np.unique(df[col])
    means = []

    for sigma in unique_sigmas:
        print("Number of elements for value", sigma, "is", len(df[df[col] == sigma]["fails"].values))
        mean = np.mean(df[df[col] == sigma]["fails"].values)
        standDev = np.std(df[df[col] == sigma]["fails"].values)
        means.append(mean)

    plt.plot(unique_sigmas, means)
    plt.title(f"Average number of fails with different {col}")
    plt.xlabel(f"Values of {col}")
    plt.ylabel("Number of fails")
    plt.show()

```
